```python
#IMPORT LIBRARY
import pandas as pd
import numpy as np
```

```python
#IMPORT DATASET

FINANCIAL_MARKET=pd.read_csv('https://raw.githubusercontent.com/YBI-Foundation/Dataset/main/Financial%20Market%20News.csv',encoding="ISO-8859-1")
```

```python
news=[]
for row in range(0,len(FINANCIAL_MARKET.index)):
    news.append(''.join(str(x) for x in FINANCIAL_MARKET.iloc[row,2:27]))
```

```python
type(news)
```

```python
news[0]
```

```python
x=news
```

```python
#get feature conversion to bag of words
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv= CountVectorizer(lowercase=True,ngram_range=(1,1))
```

```python
x=cv.fit_transform(x)
```

```python
y=FINANCIAL_MARKET['Label']
y
```

```python
#get train test split
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,stratify=y,random_state=2529)
x_train,x_test,y_train,y_test
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(n_estimators=200)
```

```python
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
confusion_matrix(y_test,y_pred)
```

```python
classification_report(y_test,y_pred)
```
